```python
import os
import pandas as pd
import csv

directory = "/Users/vincewu/Bees-Project"
processed_folder = "Fully Processed"
```

```python
log_folder = 'Data/DPC'

files = set()

filelist = os.listdir(os.path.join(directory, log_folder))
for filename in sorted(filelist):
    if filename.endswith('.csv') and filename[:2]=='lo':
        print(filename)
        files.add(filename[:14])

for file in sorted(files):
    print(file)
```

```python
compiled_folder = 'Data/DPC (Compiled)'

for file in sorted(files):
    with open(os.path.join(directory, compiled_folder, file+'.csv'), 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows([])

filelist = os.listdir(os.path.join(directory, log_folder))
for filename in sorted(filelist):
    if filename.endswith('.csv'):
        pathname = os.path.join(directory, log_folder, filename)
        # print(filename)
        try:
            df = pd.read_csv(pathname, engine='python',
                             sep="[;,. ]", header=None, skiprows=22)
        except pd.errors.EmptyDataError:
            df = pd.DataFrame()
        print(filename)
        if df.empty:
            print("empty")
            continue
        df.dropna(axis=1, how='all', inplace=True)

        df=df.iloc[:, [0, 1, 4, 5, 14]]
        df.columns = [i for i in range(len(df.columns))]

        # print(df)

        for i in range(len(df.iloc[:, 0])):
            date = df.iloc[i, 0]
            # print(date)
            for k in range(len(date)):
                if date[k]=='-':
                    break
                elif k==len(date)-1:
                    date_temp = []
                    date_index = 0
                    while date_index < len(date):
                        if date[date_index] >= '0' and date[date_index] <= '9':
                            for j in range(date_index+1, len(date)):
                                if date[j] < '0' or date[j] > '9':
                                    date_temp.append(date[date_index:j])
                                    date_index = j
                                    break
                                if j==len(date)-1:
                                    date_temp.append(date[date_index:])
                                    date_index = j

                        date_index += 1

                    if len(date_temp[0])==1:
                        date_temp[0]='0'+date_temp[0]
                    if len(date_temp[1])==1:
                        date_temp[1]='0'+date_temp[1]

                    date = '-'.join(
                        ['20'+date_temp[2], date_temp[0], date_temp[1]])
            # print(date)
                    
            with open(os.path.join(directory, compiled_folder, 'log_'+date.replace('-', '_')+'.csv'), 'a') as file:
                writer = csv.writer(file)
                writer.writerow(df.iloc[i, [0, 1, 3, 4]])
                
            
```

```python
log_folder = 'Data/DOC'
compiled_folder = 'Data/DOC (Compiled)'

for file in sorted(files):
     with open(os.path.join(directory, compiled_folder, file+'.csv'), 'w', newline='') as f:
          writer = csv.writer(f)
          writer.writerows([])

filelist = os.listdir(os.path.join(directory, log_folder))
for filename in sorted(filelist):
    if filename.endswith('.csv'):
        pathname = os.path.join(directory, log_folder, filename)
        print(filename)
        try:
            df = pd.read_csv(pathname, engine='python', sep="[;,. ]", header=None, skiprows=24)
        except pd.errors.EmptyDataError:
            df = pd.DataFrame()
        # print(filename)
        if df.empty:
            print("empty")
            continue

        if df.iloc[0, 0]=='Timestamp':
            df.columns=df.iloc[0]
            df=df.drop(index=0)
            df.index-=1
        df.dropna(axis=1, how='all', inplace=True)

        df = df.iloc[:, [0, 1, 4, 5, 7]]
        df.columns = [i for i in range(len(df.columns))]

        # print(df)

        # file_date=pd.to_datetime(filename[4:14].replace('_', '-'))

        for i in range(len(df.iloc[:, 0])):
            date=df.iloc[i, 0]
            
            if pd.to_datetime(date).year==2023:
                df.iloc[i, 0] = str(pd.to_datetime(df.iloc[i, 0]+' ' + df.iloc[i, 1]) +
                                    pd.Timedelta(minutes=45))[:10]
                date = df.iloc[i, 0]
                df.iloc[i, 1]=str(pd.to_datetime(df.iloc[i, 0]+' ' + df.iloc[i, 1]) +
                    pd.Timedelta(minutes=45))[-8:]
            
            with open(os.path.join(directory, compiled_folder, 'log_'+date.replace('-', '_')+'.csv'), 'a') as file:
                writer = csv.writer(file)
                writer.writerow(df.iloc[i, [0, 1, 3, 4]])
```
