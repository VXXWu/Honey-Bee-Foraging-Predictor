```python
import pandas as pd
df=pd.read_csv('//Users/vincewu/Downloads/tagnumeros_2023V4032723upload.csv')

tag_types = ['New', 'NP', 'Nectar', 'Pollen', 'Drone', 'pollen']
catch=['NP', 'Nectar', 'Pollen', 'pollen']

for i in range(len(df)):
    print(df['Group'][i], i)
    for j in range(len(df['Group'][i])):
        check=False
        if (df['Group'][i][j] >= 'A' and df['Group'][i][j] <= 'Z') or (df['Group'][i][j] >= 'a' and df['Group'][i][j] <= 'z'):
            temp=str(df['Group'][i][j])
            for k in range(j+1, len(df['Group'][i])):
                if (df['Group'][i][k] >= 'A' and df['Group'][i][k] <= 'Z') or (df['Group'][i][k] >= 'a' and df['Group'][i][k] <= 'z'):
                    temp += str(df['Group'][i][k])
                    if temp in tag_types:
                        if temp in catch:
                            df['Group'][i]='Catch'
                        else:
                            df['Group'][i]=temp
                        check=True
                        break

                    # print(temp)
                else:
                    break

            if check: break
            # if temp in tag_types:
            #     if temp in catch:
            #         df['Group'][i]='Catch'
            #     else:
            #         df['Group'][i]=temp
            #     break
```
